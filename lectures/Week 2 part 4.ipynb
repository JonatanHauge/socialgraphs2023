Erdős-Rényi Networks

Consider an Erdős-Rényi network with N = 3,000 nodes, connected to each other with probability p = 10–3.

What is the expected number of links, 〈L〉?

In which regime is the network?

Calculate the probability pc so that the network is at the critical point

Given the linking probability p = 10–3, calculate the number of nodes Ncr so that the network has only one component.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

```python
n=3000
p = 0.001
G = nx.erdos_renyi_graph(n,p)
```

```python
mean_links = p*n*(n-1)/2
print('expected links:', mean_links)
mean_degree = p*(n-1)
print('mean degree: ', mean_degree)
print('log n: ', np.log(n))
# The mean degree is in the supercritival regime since 1 < k < log n

# Calculate the probability that <k>=1

p_k = scipy.special.binom(n-1, 1)*p**1 * (1-p)**(n-1-1)
print('p_k: ', p_k)
```

In order for the network to have only one connected component p has to surpass (ln n)/n.

This can be solved as 

$$ p>\frac{ln~n}{n} \iff 0 > \frac{log~n - np}{n} \iff np > log~n \iff n > \frac{log~n}{p}$$

which for p=0.001 is solved by n> 9118

```python
n = 200
k = 1.5
p = k/(n-1)

G = nx.erdos_renyi_graph(n=n, p=p, seed=None, directed=False)

fig = plt.figure("Degree of a random graph", figsize=(16, 8))
# Create a gridspec for adding subplots of different sizes
axgrid = fig.add_gridspec(1,2)
Gcc = G.subgraph(sorted(nx.connected_components(G), key=len, reverse=True)[0])

pos = nx.spring_layout(G, seed=10396953)

ax0 = fig.add_subplot(axgrid[0])
nx.draw_networkx_nodes(G, pos, ax=ax0, node_size=20)
nx.draw_networkx_edges(G, pos, ax=ax0, alpha=0.4)
nx.draw_networkx_nodes(Gcc, pos, ax=ax0, node_size=20,node_color="red")

ax1 = fig.add_subplot(axgrid[1])
nx.draw_networkx_nodes(Gcc, pos, ax=ax1, node_size=20,node_color="red")
nx.draw_networkx_edges(Gcc, pos, ax=ax1, alpha=0.4)
ax1.set_title("Connected components of G")
ax1.set_axis_off()
```

Choose a node at random from the GCC. (Hint: You may want to try random.choice.)

Find all nodes that are precisely 2 steps away from that node. (Hint. I suggest networkx.single_source_shortest_path_length)

Plot the GCC with the following choice of colors. Starting node black ("#000000"). The nodes 2 steps away red ("#ff0000"). All other nodes blue ("#A0CBE2"). Again, I suggest using networkx.draw() and reading the documentation carefully find out how to color individual nodes.

```python
node = np.random.choice(Gcc.nodes)
paths = nx.single_source_shortest_path_length(Gcc, node, cutoff=2)
two_steps = [key for key, val in paths.items() if val == 2]
print(two_steps)
```

```python
fig = plt.figure("Degree of a random graph", figsize=(16, 8))

ax1 = fig.add_subplot(axgrid[0])
ax1.set_title("Gcc")
nx.draw_networkx_nodes(Gcc, pos, ax=ax1, node_size=20,node_color="#A0CBE2", edgecolors='black', linewidths=0.5)
nx.draw_networkx_nodes(Gcc, pos, ax=ax1, nodelist = [node], node_size=20,node_color="#000000", edgecolors='black', linewidths=0.5)
nx.draw_networkx_nodes(Gcc, pos, ax=ax1, nodelist = two_steps, node_size=20,node_color="#ff0000", edgecolors='black', linewidths=0.5)
nx.draw_networkx_edges(Gcc, pos, ax=ax1, alpha=0.4)
plt.show()
```
