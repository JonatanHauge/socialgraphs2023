> Finally, we can create our own corpus (see Ch. 2 Sec. 1.9 for more details):
>    * create a list of file names `file_list` and order them with the same order in `df_rappers`, e.g., 03_Greedo.txt first, 22Gz.txt second, etc.
>    * use the function `PlaintextCorpusReader` from `nltk.corpus` to create the rapper corpus.
>    * use the function `nltk.Text()` as follows `nltk.Text(YOUR_CORPUS_HERE.words())`.
>
> Now you can use this nltk Text object as any other you've seen in the examples of Chapter 1!!

```python
import pandas as pd
import requests
import re
import numpy as np

baseurl = "http://en.wikipedia.org/w/api.php?"
df_rappers = pd.read_csv(r"..\files\Rappers.csv")
file_names = []

for name in df_rappers['WikipediaPageName']:
    params = {"action": "query",
    "prop": "extracts",
    "exlimit":"1",
    "explaintext": "1",
    "format": "json",
    "titles": name}
    
    wikitext = requests.get(baseurl, params = params)
    wikijson = wikitext.json()
    ID = list(wikijson['query']['pages'].keys())[0]
    text = wikijson['query']['pages'][ID]['extract']
    spec_char = '?\/:*<>' 
    for c in spec_char:
        name = name.replace(c, "") #Remove special characters from name
    
    path = 'rappers_text/{}.txt'.format(name)
    with open(path, 'w', encoding = 'utf-8') as f:
        f.write(text)
    file_names.append('{}.txt'.format(name))
```

```python
from nltk.corpus import PlaintextCorpusReader

tmp = PlaintextCorpusReader('rappers_text/', '.*\.txt') # .*\.txt means open all files in directory
text = nltk.Text(tmp.words())
len(text)
```

â¼ Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?

```python
from nltk.corpus import inaugural
import matplotlib.pyplot as plt

year, men, women, people = [],[],[],[]

for id in inaugural.fileids():
    tmp = [w.lower() for w in nltk.Text(inaugural.words(id))]
    year.append(int(id[:4]))
    men.append(tmp.count('men'))
    women.append(tmp.count('women'))
    people.append(tmp.count('people'))
   
fig, ax = plt.subplots(1,1,figsize=(15,5))

ax.plot(year, men, c = 'blue')
ax.plot(year, women, c = 'red')
ax.plot(year, people, c = 'green')
```

The usage of the word 'woman' does not happen until sometime in the 1900's, crazy!

```python
from nltk.corpus import brown
cfd = nltk.ConditionalFreqDist(
           (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['car', 'you', 'government', 'alien', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

**2.8.23**

Zipf's Law: Let f(w) be the frequency of a word w in free text. Suppose that all the words of a text are ranked according to their frequency, with the most frequent word first. Zipf's law states that the frequency of a word type is inversely proportional to its rank (i.e. f Ã r = k, for some constant k). For example, the 50th most common word type should occur three times as frequently as the 150th most common word type.

* Write a function to process a large text and plot word frequency against word rank using pylab.plot. Do you confirm Zipf's law? (Hint: it helps to use a logarithmic scale). What is going on at the extreme ends of the plotted line?

* Generate random text, e.g., using random.choice("abcdefg "), taking care to include the space character. You will need to import random first. Use the string concatenation operator to accumulate characters into a (very) long string. Then tokenize this string, and generate the Zipf plot as before, and compare the two plots. What do you make of Zipf's Law in the light of this?

```python
# (a)
def plot_zip(text):
    l = len(text)
    count = sorted([val/l for (key, val) in FreqDist(text).items()], reverse = True)
    print(count[:5])
    
    fig, ax = plt.subplots(1,1, figsize=(5,5))
    ax.plot(range(0,len(count)), count)
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_xlabel('log-rank')
    ax.set_ylabel('frequency')
    
plot_zip(brown_text)
```

It is a powerlaw, but in both ends of the line it seems to deviate from the 'straight' line. In the beginning it deviates, which makes sense since it would be weird if the most frequent word was for instance 3 times as frequent as the 3. most frequent word. The most frequent words are often words like 'a', 'he', and 'it', which occur almost equally many times (Try and show this in code).

In the other end we see the effect of unique words only used once, which there are several of and therefore it deviatea from the power law in the end also.
